```python
import numpy as np
import pandas as pd
```

### EX1

```python
authors = pd.DataFrame({'author_id' : [1, 2, 3],
                        'author_name' : ['Тургенев', 'Чехов', 'Островский']}, columns=['author_id', 'author_name'])
authors
```

```python
book = pd.DataFrame({'author_id' : [1, 1, 1, 2, 2, 3, 3],
                     'book_title' : ['Отцы и дети', 'Рудин', 'Дворянское гнездо', 'Толстый и тонкий', 'Дама с собачкой', 'Гроза', 'Таланты и поклонники'],
                     'price' : [450, 300, 350, 500, 450, 370, 290]})
book
```

### EX2

```python
authors_price = pd.merge(authors, book, on = 'author_id', how = 'outer')
authors_price
```

### EX3

```python
authors_price.nlargest(5, 'price')
```

## EX4

```python
authors_stat = pd.concat([authors_price.groupby('author_name')['price'].max().rename('max_price'), authors_price.groupby('author_name')['price'].min().rename('min_price'), authors_price.groupby('author_name')['price'].mean().rename('mean_price')], axis=1, ignore_index=False)
authors_stat
```

```python
authors_price.insert(3, 'cover', ['твердая', 'мягкая', 'мягкая', 'твердая', 'твердая', 'мягкая', 'мягкая'])
authors_price
```
