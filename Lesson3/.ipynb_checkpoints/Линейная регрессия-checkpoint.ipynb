```python
import numpy as np
import pandas as pd
```

#### Работа со встроенными наборами данных в scikit-learn

Набор данных Boston House Prices

```python
from sklearn.datasets import load_boston
```

```python
boston = load_boston()
```

```python
boston.keys()
```

Данные о недвижимости

```python
data = boston.data
```

```python
data.shape
```

```python
data
```

Target - величина, которую требуется предсказать (цена на недвижимость)

```python
target = boston.target
```

Названия признаков

```python
feature_names = boston.feature_names
```

```python
feature_names
```

Описание датасета

```python
for line in boston.DESCR.split('\n'):
    print(line)
```

создадим два датафрейма


```python
x = pd.DataFrame(data, columns=feature_names)
```

```python
x.head()
```

```python
x.shape
```

```python
x.info
```

```python
y = pd.DataFrame(target, columns=['price'])
```

```python
y.head()
```

```python
y.info
```

#### Разбиение данных на тренировочный и тестовый датасеты

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=42)
```

#### Построение модели линейной регрессии

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

Задача линейной регресии - найти подходящие коэффициенты w

при признаках х для вычислений целевой переменной у,

минимизируя ошибку е:

y = w<sub>0</sub> + w<sub>1</sub> ∗  x<sub>1</sub> + w<sub>2</sub> ∗ x<sub>2</sub>+...+w<sub>m</sub> ∗ x<sub>m</sub> + e

```python
lr.fit(x_train, y_train)
```

```python
y_pred = lr.predict(x_test)
```

```python
check_test = pd.DataFrame({'y_test': y_test['price'],
                           'y_pred': y_pred.flatten()},
                          columns=['y_test', 'y_pred'])
```

```python
check_test.head(10)
```

#### Метрики оценки качества регрессионных моделей

Средняя квадратичная ошибка

```python
check_test['error'] = check_test['y_pred'] - check_test['y_test']
```

```python
check_test.head()

```

```python
initial_mse = (check_test['error'] ** 2).mean()
initial_mse
```

```python
from sklearn.metrics import mean_squared_error
```

```python
initial_mse = mean_squared_error(y_test, y_pred)
initial_mse
```

Средняя абсолютная ошибка

```python
(np.abs(check_test['error'])).mean()
```

```python
from sklearn.metrics import mean_absolute_error
```

```python
mean_absolute_error(y_test, y_pred)
```

R<sup>2</sup>

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test, y_pred)
```

#### Просмотр коэффициентов линейной регрессии

```python
lr.intercept_
```

w<sub>1</sub>...w<sub>m</sub>

```python
lr.coef_
```

```python
feats = ['CRIM', 'ZN', 'CHAS', 'NOX', 'RM', 'DIS', 'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT']
```

```python
def create_model(x_train, y_train, x_test, y_test, feats, model):
    model.fit(x_train.loc[:, feats], y_train)
    y_predict = model.predict(x_test.loc[:, feats])
    mse = mean_squared_error(y_test, y_predict)
    return mse
```

```python
create_model(x_train_scaled, y_train, x_test_scaled, y_test, feats, LinearRegression())
```

#### Модели линейной регрессии с регуляризацией

Lasso (линейная регрессия с L1-регуляризацией)

```python
# Параметр alpha отвечает за легуляризацию
model = Lasso(alpha=0.003)
```

```python
create_model(x_train_scaled, y_train, x_test_scaled, y_test, feats, model)
```

Ridge (линейная регрессия с L2-регуляризацией)

```python
# Параметр alpha отвечает за легуляризацию
model = Ridge(alpha=0.001)
```

```python
create_model(x_train_scaled, y_train, x_test_scaled, y_test, feats, model)

```

```python
n = 21
coeffs = np.zeros((n, len(feats)))
alpha_list = np.logspace(-3, 1, n)
for i, val in enumerate(alpha_list):
    lasso = Lasso(alpha=val)
    lasso.fit(x_train_scaled.loc[:, feats], y_train)
    coeffs[i, :] = lasso.coef_.flatten()

for i in range(len(feats)):
    plt.plot(alpha_list, coeffs[:, i])
plt.title('Убавние абсолютных значений весов признаков\n при увеличении коэффициента регуляризации alpha (Lasso)')
plt.xlabel('alpha')
plt.ylabel('Вес признака')
plt.legend(feats)
```

```python
n = 66
coeffs = np.zeros((n, len(feats)))
alpha_list = np.logspace(-3, 3.5, n)
for i, val in enumerate(alpha_list):
    ridge = Ridge(alpha=val)
    ridge.fit(x_train_scaled.loc[:, feats], y_train)
    coeffs[i, :] = ridge.coef_.flatten()

for i in range(len(feats)):
    plt.plot(alpha_list, coeffs[:, i])
plt.title('Убавние абсолютных значений весов признаков\n при увеличении коэффициента регуляризации alpha (Ridge)')
plt.xlabel('alpha')
plt.ylabel('Вес признака')
plt.legend(feats)
```
